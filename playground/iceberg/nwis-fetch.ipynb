```python
import dataretrieval.nwis as nwis
import hvplot.pandas # noqa: E402
```

```python
nwis_df = nwis.get_record(sites='14182500', service='iv', start='2025-01-01', end='2025-04-01')
nwis_df
```

```python
nwis_df.hvplot.line(
    x='datetime',
    y='00060',
    title='USGS Streamflow Instantaneous Values for Site 14182500',
    xlabel='Date',
    ylabel='Streamflow (m^3/s)',
    width=800,
    height=400
)
```

```python
import pandera as pa
```

```python
schema = pa.DataFrameSchema({
    "location_id": pa.Column(pa.String, nullable=False, coerce=True),
    "value_time": pa.Column(pa.DateTime, nullable=False, coerce=True),
    "value": pa.Column(pa.Float64, nullable=False, coerce=True)
},
strict="filter",
drop_invalid_rows=True
)
```

```python
schema.validate(nwis_df, lazy=True)
```

```python
df = nwis_df.reset_index()

df = df.rename(columns={
    '00060': 'value',
    'datetime': 'value_time',
    'site_no': 'location_id'
})
df.value = df.value.astype('float64')
```

```python
schema.validate(df, lazy=True)
```
